# Conversão de Imagem

<p style="text-align: justify;">
A conversão de uma imagem do espaço de cores RGB (Red, Green, Blue) para tons de cinza envolve a redução da informação de cor para uma única dimensão, representando a intensidade luminosa de cada pixel. Em uma imagem em tons de cinza, a ausência de cor é representada por diferentes tons de cinza, onde 0 é preto e 255 é branco. 

A fórmula comum para converter uma imagem RGB para tons de cinza é uma média ponderada das componentes de cor R, G e B. A ponderação reflete a sensibilidade do olho humano às diferentes cores. A fórmula geral para a conversão é:
</p>

<p style="text-align: center"> <b>Tons de Cinza=0.299 × R + 0.587 × G + 0.114 × B </b></p>

Onde:

- R,G,B são as intensidades das componentes vermelha, verde e azul, respectivamente.
- Os coeficientes (0.299, 0.587, 0.114) refletem as ponderações.

<p style="text-align: justify;">
 Podemos observar a seguir duas formas de converter as imagens de RGB para Tons de Cinza. A primeira é utilizando a fórmula e a segunda utilizando uma função do opencv chamda de cvtColor().
</p>

```python
import cv2
import matplotlib.pyplot as plt

# Carregar a imagem RGB
imagem_rgb = cv2.imread('Imagens-VC/exemplo-imagem.jpg')
imagem_rgb = cv2.cvtColor(imagem_rgb, cv2.COLOR_BGR2RGB)

# Extrair as componentes R, G, B
R = imagem_rgb[:, :, 0]
G = imagem_rgb[:, :, 1]
B = imagem_rgb[:, :, 2]

# Calcular a intensidade em tons de cinza usando a fórmula
imagem_tons_de_cinza = 0.299 * R + 0.587 * G + 0.114 * B
```

# Suavização - Filtro Gaussiano

<p style="text-align: justify;">
A suavização de imagens, também conhecida como filtragem, é um processo no qual o objetivo é reduzir o ruído ou as variações abruptas de intensidade na imagem, tornando-a mais homogênea. Este processo é comumente realizado através da aplicação de filtros.

Um dos filtros mais utilizados para suavização é o Filtro Gaussiano. Esse filtro é baseado na distribuição Gaussiana e é eficaz para reduzir ruídos enquanto mantém as bordas da imagem. A aplicação do filtro Gaussiano envolve convolução da imagem original com o kernel (máscara) Gaussiano.
</p>

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Carregar a imagem
imagem = cv2.imread('Imagens-VC/ruido2.png')
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

# Aplicar suavização com filtro Gaussiano
imagem_suavizada = cv2.GaussianBlur(imagem_rgb, (5, 5),0) # Função para aplicar o Gaussiano. (5,5) é o tamanho do kernel
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(imagem_rgb)
plt.title('Imagem Original')

plt.subplot(1, 2, 2)
plt.imshow(imagem_suavizada)
plt.title('Imagem Suavizada (Filtro Gaussiano)')

plt.show()
```

# Suavização - Filtro Mediana

<p style="text-align: justify;">
O filtro de mediana é uma técnica de suavização de imagens que visa reduzir o ruído, preservando as bordas. Ao contrário de outros filtros, como o filtro média que utiliza a média dos valores dos pixels em uma vizinhança, <b>o filtro de mediana substitui o valor de cada pixel pela mediana dos valores em sua vizinhança. </b>

A mediana é o valor central em um conjunto ordenado de dados. Essa propriedade torna o filtro de mediana robusto contra outliers e adequado para preservar bordas em uma imagem.

A aplicação prática do filtro de mediana envolve a criação de uma "janela" que percorre toda a imagem, e para cada posição da janela, o valor do pixel é substituído pela mediana dos valores dentro dessa janela.
</p>

```python
import cv2
import matplotlib.pyplot as plt

# Carregar a imagem
imagem = cv2.imread('Imagens-VC/pulmao.jpg')
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

# Aplicar o filtro de mediana
imagem_suavizada = cv2.medianBlur(imagem_rgb, 5)  # O segundo parâmetro é o tamanho da janela (ímpar)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(imagem_rgb)
plt.title('Imagem Original')

plt.subplot(1, 2, 2)
plt.imshow(imagem_suavizada)
plt.title('Imagem Suavizada (Filtro de Mediana)')

plt.show()
```

# Histograma

<p style="text-align: justify;">
Um histograma em visão computacional é uma <b>representação gráfica da distribuição de intensidades de cores ou níveis de cinza </b> em uma imagem digital. Ele é uma ferramenta crucial para entender e analisar a composição de uma imagem.
    
O histograma é calculado contando a frequência de ocorrência de cada valor de intensidade possível na imagem. Aqui está uma representação básica da fórmula para calcular o histograma de uma imagem em tons de cinza:
</p>

# Histograma

<p style="text-align: justify;">
Seja I a imagem em tons de cinaz e n o número total de pixels na imagem. 
    
O histograma H(i) para o valor de intensidade i dado por:
    
    H(i) = Ni / n
    
Onde:
    
    - Ni é o número de pixels na imagem que possuem intensidade i. 
    - n é o número total de pixels na imagem.
    
Para uma imagem colorida, cada canal de cor (vermelho, verde, azul) terá seu próprio histograma, mas o conceito básico é o mesmo.
</p>

```python
import cv2
import matplotlib.pyplot as plt

# Carregar a imagem em escala de cinza
imagem = cv2.imread('Imagens-VC/ruido4.jpg', cv2.IMREAD_GRAYSCALE)

histograma_original = cv2.calcHist([imagem], [0], None, [256], [0, 256])
```

```python
# Exibir as imagens
fig, axs = plt.subplots(1, 2, figsize=(16, 5))

# Exibir as imagens e os histogramas
axs[0].imshow(imagem, cmap='gray')
axs[0].set_title('Imagem Original')

axs[1].plot(histograma_original, color='black')
axs[1].set_title('Histograma Original')

# Ajustar o layout
plt.tight_layout()
plt.show()
```

# Equalização de Histograma

<p style="text-align: justify;">
A equalização de histograma é uma técnica de processamento de imagens que visa melhorar o contraste ao redistribuir as intensidades dos pixels em uma imagem. A ideia é transformar o histograma da imagem original de modo a obter uma distribuição mais uniforme das intensidades.

O histograma de uma imagem representa a frequência de ocorrência de diferentes intensidades de pixels. Na equalização de histograma, a transformação é realizada para que o histograma resultante seja o mais uniforme possível, buscando espalhar as intensidades ao longo de todo o intervalo disponível.
</p>

# Equalização de Histograma

<p style="text-align: justify;">
A equalização de histograma envolve os seguintes passos: </p>
    
    - Calcular o histograma da imagem original.
    - Calcular a função de distribuição acumulada (CDF - Cumulative Distribution Function) do histograma.
    - Mapear cada valor de intensidade na imagem original para um novo valor usando a CDF.


```python
import cv2
import matplotlib.pyplot as plt

# Carregar a imagem em escala de cinza
imagem = cv2.imread('Imagens-VC/ruido4.jpg', cv2.IMREAD_GRAYSCALE)

# Aplicar equalização de histograma
imagem_equalizada = cv2.equalizeHist(imagem)
```

```python
# Exibir as imagens
fig, axs = plt.subplots(1, 4, figsize=(16, 5))

# Exibir as imagens e os histogramas
axs[0].imshow(imagem, cmap='gray')
axs[0].set_title('Imagem Original')

axs[1].plot(histograma_original, color='black')
axs[1].set_title('Histograma Original')

axs[2].imshow(imagem_equalizada, cmap='gray')
axs[2].set_title('Imagem Equalizada')

axs[3].plot(histograma_equalizado, color='black')
axs[3].set_title('Histograma Equalizado')

# Ajustar o layout
plt.tight_layout()
plt.show()
```

# Contrast Limited Adaptive Histogram Equalization (CLAHE)

<p style="text-align: justify;">
O CLAHE é uma técnica de processamento de imagem que <b> estende a equalização de histograma para melhorar o contraste em imagens, especialmente em regiões locais </b>. Ao contrário da equalização de histograma global, <b> o CLAHE opera em pequenas regiões da imagem </b>, adaptando-se localmente às variações de contraste.

O CLAHE divide a imagem em "blocos" ou "tiles" e aplica a equalização de histograma a cada bloco individualmente. No entanto, para evitar amplificar o ruído em áreas com baixo contraste, é introduzido um limite de contraste. Se o histograma de um bloco ultrapassar esse limite, ele é redistribuído de maneira uniforme, limitando assim o aumento do contraste.
    
</p>


# Contrast Limited Adaptive Histogram Equalization (CLAHE)

Algumas características podem ser observadas a seguir:

- Divisão em Blocos (Tiles): A imagem é dividida em blocos ou tiles. Cada bloco é tratado individualmente para garantir que as transformações sejam aplicadas localmente.
- Equalização de Histograma Local: Para cada bloco, a equalização de histograma é aplicada. Isso significa que o histograma de intensidades dentro de cada bloco é modificado para ser mais uniforme, aumentando o contraste local.
- Limitação do Contraste (ClipLimit): Para evitar amplificar o ruído em áreas com baixo contraste, é introduzido um limite de contraste chamado clipLimit. Se o histograma do bloco ultrapassar esse limite, os pixels são redistribuídos de maneira uniforme para evitar a amplificação excessiva do contraste.

# Contrast Limited Adaptive Histogram Equalization (CLAHE)

<p style="text-align: justify;">

O parâmetro clipLimit no CLAHE (Contrast Limited Adaptive Histogram Equalization) em OpenCV define o limite de contraste, controlando o quanto o contraste pode ser amplificado localmente em cada bloco. O intervalo típico para o clipLimit é de 0 a 40.

</p>

- Valor Baixo (perto de 0): Limita significativamente a amplificação do contraste local, resultando em uma equalização mais suave. Pode preservar detalhes finos, mas o contraste global pode não ser tão melhorado.

- Valor Alto (perto de 40): Permite uma amplificação mais agressiva do contraste local. Pode resultar em maior realce de bordas e detalhes locais, mas também pode aumentar a sensibilidade ao ruído, levando a artefatos indesejados.

<p style="text-align: justify;">
O ajuste do clipLimit é um processo empírico e depende das características específicas da imagem e dos requisitos da aplicação. Experimentar diferentes valores do clipLimit é uma boa prática para encontrar o equilíbrio adequado entre realce de contraste e preservação de detalhes, enquanto evita artefatos indesejados.
</p>

```python
import cv2
import matplotlib.pyplot as plt

# Carregar a imagem em escala de cinza
imagem = cv2.imread('Imagens-VC/ruido4.jpg', cv2.IMREAD_GRAYSCALE)

# Configurar o objeto CLAHE
clahe = cv2.createCLAHE(clipLimit=20.0, tileGridSize=(8, 8))

# Aplicar CLAHE
imagem_clahe = clahe.apply(imagem)

# Calcular o histograma da imagem original
histograma_original = cv2.calcHist([imagem], [0], None, [256], [0, 256])

# Calcular o histograma da imagem com CLAHE
histograma_clahe = cv2.calcHist([imagem_clahe], [0], None, [256], [0, 256])
```

```python
fig, axs = plt.subplots(1, 4, figsize=(16, 5))

# Exibir as imagens e os histogramas
axs[0].imshow(imagem, cmap='gray')
axs[0].set_title('Imagem Original')

axs[1].plot(histograma_original, color='black')
axs[1].set_title('Histograma Original')

axs[2].imshow(imagem_clahe, cmap='gray')
axs[2].set_title('Imagem CLAHE')

axs[3].plot(histograma_clahe, color='black')
axs[3].set_title('Histograma CLAHE')

# Ajustar o layout
plt.tight_layout()
plt.show()
```

# Métricas

Na visão computacional, várias métricas são utilizadas para avaliar a qualidade de uma imagem em relação a uma imagem de referência ou para comparar duas imagens. Algumas das métricas mais comuns incluem:
- RMSE (Root Mean Squared Error), 
- AMBE (Average Mean Absolute Error), 
- PSNR (Peak Signal-to-Noise Ratio) e 
- SSIM (Structural Similarity Index).

# Métricas - RMSE (Root Mean Squared Error)

O RMSE é uma medida da diferença entre os valores previstos por um modelo ou processo e os valores observados ou reais. Na comparação de imagens, o RMSE é calculado como a raiz quadrada da média dos quadrados das diferenças pixel a pixel entre as imagens. É dado pela seguinte fórmula:



Onde: 
- N é o número total de pixels na imagem.
- I1(i) e I2(i) são valores de intensidade dos pixels nas imagens de comparação

```python
import cv2
import numpy as np

def calcular_rmse(imagem1, imagem2):
    # Verificar se as imagens têm as mesmas dimensões
    if imagem1.shape != imagem2.shape:
        raise ValueError("As imagens têm dimensões diferentes.")

    # Calcular o erro quadrático entre as imagens
    erro_quadratico = np.square(imagem1.astype(np.float32) - imagem2.astype(np.float32))

    # Calcular o RMSE
    rmse = np.sqrt(np.mean(erro_quadratico))

    return rmse

# Exemplo de uso
imagem1 = cv2.imread('Imagens-VC/ruido5.jpg', cv2.IMREAD_GRAYSCALE)
imagem2 = cv2.imread('Imagens-VC/imagem_clahe.jpg', cv2.IMREAD_GRAYSCALE)

rmse = calcular_rmse(imagem1, imagem2)
print("RMSE entre as imagens:", rmse)
```

# Métricas - AMBE (Average Mean Absolute Error)

A AMBE é uma medida da média das diferenças absolutas entre os valores de intensidade dos pixels nas imagens de comparação. É calculada pela média das diferenças absolutas entre os pixels. A fórmula é a seguinte:



```python
import cv2
import numpy as np

def calcular_ambe(imagem1, imagem2):
    # Verificar se as imagens têm as mesmas dimensões
    if imagem1.shape != imagem2.shape:
        raise ValueError("As imagens têm dimensões diferentes.")

    # Calcular a diferença absoluta entre as imagens
    diferenca_absoluta = np.abs(imagem1.astype(np.float32) - imagem2.astype(np.float32))

    # Calcular o AMBE (Average Mean Absolute Error)
    ambe = np.mean(diferenca_absoluta)

    return ambe

# Exemplo de uso
imagem1 = cv2.imread('Imagens-VC/ruido5.jpg', cv2.IMREAD_GRAYSCALE)
imagem2 = cv2.imread('Imagens-VC/imagem_clahe.jpg', cv2.IMREAD_GRAYSCALE)

ambe = calcular_ambe(imagem1, imagem2)
print("AMBE entre as imagens:", ambe)
```

# Métricas - PSNR (Peak Signal-to-Noise Ratio)

O PSNR é uma medida de qualidade de imagem que compara a qualidade de uma imagem original com uma imagem distorcida. Ele é baseado na relação entre a potência do sinal (a intensidade máxima possível do sinal) e a potência do ruído (a diferença entre o sinal original e a imagem distorcida).

O PSNR é uma métrica comum usada para avaliar a qualidade de reconstrução de imagens, onde a imagem original é comparada com uma versão reconstruída ou comprimida da mesma imagem.

A fórmula básica para calcular o PSNR é:



- MAX é o valor máximo possível para a intensidade do pixel.
- MSE é o erro médio quadrático

Um valor alto de PSNR indica que a imagem é mais semelhante à imagem original ou de referência, enquanto um valor baixo indica que a imagem é significativamente diferente da imagem de referência. 
 

```python
import cv2
import numpy as np

def calcular_psnr(imagem1, imagem2):
    # Verificar se as imagens têm as mesmas dimensões
    if imagem1.shape != imagem2.shape:
        raise ValueError("As imagens têm dimensões diferentes.")

    # Calcular o erro quadrático entre as imagens
    mse = np.mean((imagem1 - imagem2) ** 2)
    
    # Calcular o valor maximo para a intensidade do pixel
    max_pixel = np.max(imagem1)
    
    # Calcular o PSNR
    
    psnr = 10 * np.log10((max_pixel ** 2) / mse)
    
    return psnr

# Exemplo de uso
imagem1 = cv2.imread('Imagens-VC/ruido5.jpg', cv2.IMREAD_GRAYSCALE)
imagem2 = cv2.imread('Imagens-VC/imagem_clahe.jpg', cv2.IMREAD_GRAYSCALE)

psnr = calcular_psnr(imagem1, imagem2)
print("PSNR entre as imagens:", psnr)
```

# Métricas - SSIM (Structural Similarity Index)

O SSIM é uma métrica que avalia a similaridade estrutural entre duas imagens. <b>Ele mede a semelhança de padrões estruturais locais entre as imagens, levando em consideração luminância, contraste e estrutura. </b>

O SSIM varia de -1 a 1, onde 1 indica uma correspondência perfeita entre as imagens. A fórmula para o cálculo do SSIM é bastante complexa e envolve comparações locais entre janelas de pixels nas imagens.

```python
import cv2 
from skimage.metrics import structural_similarity as ssim

imagem1 = cv2.imread('Imagens-VC/ruido5.jpg', cv2.IMREAD_GRAYSCALE)
imagem2 = cv2.imread('Imagens-VC/imagem_clahe.jpg', cv2.IMREAD_GRAYSCALE)

ssim_value = ssim(imagem1, imagem2, data_range=imagem2.max() - imagem2.min())
print("SSIM entre as imagens:", ssim_value)
```
